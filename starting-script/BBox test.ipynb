```python
import json
from math import floor

import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/Users/oliveredholm/Downloads/tmp 21/www_youtube_com-1570172362477/screenshot.jpg')
```

```python
bbox = json.loads('{\"x\":240,\"y\":56,\"width\":1680,\"height\":443,\"top\":56,\"right\":1920,\"bottom\":499,\"left\":240}')
```

```python
bbox = {k: floor(v) for k, v in bbox.items()}
```

```python
scroll_info = json.loads('{\"windowX\":0,\"windowY\":0}')
```

```python
scroll_info = {k: floor(v) for k, v in scroll_info.items()}
```

```python
img = cv2.rectangle(img, (bbox['left'] + scroll_info['windowX'], bbox['top'] + scroll_info['windowY']), (bbox['right'] + scroll_info['windowX'], bbox['bottom'] + scroll_info['windowY']), (255, 0, 0), 4)
```

```python
plt.figure(figsize=(50, 50))
plt.imshow(img)
plt.show()
```
